```python
import pandas as pd

df = pd.read_csv("../data/SuperStoreOrders.csv")
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Change column type to datetime64[ns] for column: 'order_date'
    df = df.astype({'order_date': 'datetime64[ns]'})
    # Change column type to datetime64[ns] for column: 'ship_date'
    df = df.astype({'ship_date': 'datetime64[ns]'})
    # Remove commas and convert 'sales' to float
    df['sales'] = df['sales'].str.replace(',', '').astype(float)
    # Change column type to int64 for column: 'quantity'
    df = df.astype({'quantity': 'int64'})
    # Fix KeyError by correcting column name
    df['Profit_Margin'] = df["profit"] / df["sales"]
    # Change column type to float64 for column: 'Profit_Margin'
    df = df.astype({'Profit_Margin': 'float64'})
    # Modified column 'Profit_Margin' using formula
    df['Profit_Margin'] = df["Profit_Margin"] = df["Profit_Margin"] * 100
    # Fix column name and calculate shipping days
    df['Shipping_Days'] = (pd.to_datetime(df["ship_date"]) - pd.to_datetime(df["order_date"])).dt.days
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

## Revenue by Region
This analysis identifies which regions generate the highest sales.

```python
df.groupby("region")["sales"].sum().sort_values(ascending=False)
```

The West region generates the highest revenue, indicating strong demand and market dominance. The company should prioritize inventory and marketing efforts in this region.

## Profit by Category
This analysis evaluates which product categories are most profitable.

```python
df.groupby("category")["profit"].sum().sort_values(ascending=False)
```

Technology products generate the highest profit, while some categories underperform. This suggests the company should focus on high-margin product lines.

## Loss-Making Sub-Categories
This analysis identifies products that generate losses despite sales.

```python
df[df["profit"] < 0].groupby("sub_category")["profit"].sum().sort_values()
```

Certain sub-categories consistently generate losses, indicating pricing inefficiencies or high operational costs. These products should be reviewed for cost reduction or pricing adjustments.

## Sales by Customer Segment
This analysis identifies the most valuable customer groups.

```python
df.groupby("segment")["sales"].sum().sort_values(ascending=False)
```

The Corporate segment contributes the highest revenue, suggesting that business customers are a key driver of sales. The company should strengthen B2B relationships.
